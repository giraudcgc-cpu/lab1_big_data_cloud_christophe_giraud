```python
%sql
--- fact_marathon ---
CREATE OR REPLACE TABLE marathos_cat.gold.fct_marathon AS
SELECT
    athlete_id,
    event_id,
    date_id,
    athlete_performance,
    athlete_performance_value,
    performance_unit,
    athlete_average_speed,
    event_number_of_finishers
FROM marathos_cat.silver.marathos_obt;

SELECT * FROM marathos_cat.gold.fct_marathon LIMIT 1;
```

```python
    %sql
    --- dim athlete ---

    CREATE OR REPLACE TABLE marathos_cat.gold.dim_athlete AS
    SELECT DISTINCT
        athlete_id,
        athlete_gender,
        athlete_year_of_birth,
        athlete_country,
        athlete_club,
        athlete_age_category
    FROM marathos_cat.silver.marathos_obt;

    SELECT * FROM marathos_cat.gold.dim_athlete LIMIT 1;
```

```python
%sql
--- dim event ---
CREATE OR REPLACE TABLE marathos_cat.gold.dim_event AS
SELECT DISTINCT
    event_id,
    event_name,
    event_type,
    unit_measure,
    unit_value,
    event_distance_or_length
FROM marathos_cat.silver.marathos_obt;

SELECT * FROM marathos_cat.gold.dim_event LIMIT 1;
```

```python
%sql
--- dim date ---
CREATE OR REPLACE TABLE marathos_cat.gold.dim_date AS
SELECT DISTINCT
    date_id,
    year_of_event,
    event_dates
FROM marathos_cat.silver.marathos_obt;

SELECT * FROM marathos_cat.gold.dim_date LIMIT 1;
```
