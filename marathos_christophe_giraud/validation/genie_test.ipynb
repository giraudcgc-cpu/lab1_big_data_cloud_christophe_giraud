```python
%sql
--- Average speed for men and women  ---
SELECT 
    a.athlete_gender,
    ROUND(AVG(f.athlete_average_speed), 2) AS avg_speed_kmh
FROM marathos_cat.gold.fct_result f
LEFT JOIN marathos_cat.gold.dim_athlete a ON f.athlete_id = a.athlete_id
WHERE a.athlete_gender IS NOT NULL
GROUP BY a.athlete_gender
    
```

```python
%sql
--- Number of distinct athletes ---
SELECT 
    COUNT(DISTINCT athlete_id) AS total_athletes
FROM 
    marathos_cat.gold.fct_result;
```

```python
%sql
--- Longest race in km ---
SELECT 
    e.event_name,
    e.unit_value AS distance_km
FROM marathos_cat.gold.dim_event e
WHERE e.event_type = 'distance'
ORDER BY e.unit_value DESC
LIMIT 1;
```

```python
%sql
--- first and last marathon dates ---
SELECT 
    MIN(event_dates) AS first_race,
    MAX(event_dates) AS last_race
FROM marathos_cat.gold.fct_result
```

```python
%sql
SELECT
    a.athlete_country, 
    ROUND(AVG(f.athlete_average_speed), 2) AS avg_speed_kmh
FROM marathos_cat.gold.fct_result f
LEFT JOIN marathos_cat.gold.dim_athlete a ON f.athlete_id = a.athlete_id
WHERE a.athlete_country IN ('FRA', 'USA', 'SWE')
GROUP BY a.athlete_country
ORDER BY a.athlete_country;
```
