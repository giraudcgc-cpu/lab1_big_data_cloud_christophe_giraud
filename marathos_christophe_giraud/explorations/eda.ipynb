# EDA for Marathos' data

```python

VOLUME_PATH = "/Volumes/marathos_cat/default/raw"
file = spark.sql(f"LIST '{VOLUME_PATH}'").display()
```

```python
df = spark.read.format("csv").option("header", True).option("inferSchema", True).load(VOLUME_PATH)
df.show(3)
```

```python
display(df.limit(3))
```

```python
print(f"They are {df.count()} rows and {len(df.columns)} columns")
print(df.dtypes)
```

```python
df.printSchema()
```

```python
# --- Using .describe() and .display() to see the results ---
df.describe().display()
```

```python
# --- Using .summary() and .display() to see the results ---
df.summary().display()
```

```python
from pyspark.sql.functions import col, sum as spark_sum

# --- In each column, when value is null, True is converted into 1 and the 1 are added ---
null_counts = df.select(
    [spark_sum(col(c).isNull().cast("int")).alias(c) for c in df.columns])

null_counts = null_counts.collect()[0].asDict()                # .collect() does the math + convert into dict
[(c, nulls) for c, nulls in null_counts.items() if nulls > 0]  # looping in the dict and only count if null
```
